### 5.4 | Treat Special Formats in Columns

```python

# experience Column 
df_encoded['experience'] = df_encoded['experience'].replace({
    '>20': 21,
    '<1': 0
})
df_encoded['experience'] = pd.to_numeric(df_encoded['experience'], errors='coerce')  # Converts to float, keeps NaNs

# last_new_job Column 
df_encoded['last_new_job'] = df_encoded['last_new_job'].replace({
    '>4': 5,
    'never': 0
})
df_encoded['last_new_job'] = pd.to_numeric(df_encoded['last_new_job'], errors='coerce')

# company_size Column 
# Mapping ranges to approximate average values
company_size_map = {
    '<10': 5,
    '10/49': 30,
    '50-99': 75,
    '100-500': 300,
    '500-999': 750,
    '1000-4999': 3000,
    '5000-9999': 7500,
    '10000+': 10000,
}

df_encoded['company_size'] = df_encoded['company_size'].map(company_size_map)
```

### 5.6 | Feature Engineering based on Error Analysis

Based on error analysis, new features were created to improve model performance:
- `training_hours_log`: log transform to reduce skewness.
- `devindex_training_interaction`: feature interaction to capture combined effect.
- `company_size_grouped`: grouped rare company sizes.
- `training_hours_bin`: binned version of training hours (categorical).
- `experience_level`: grouped experience into ranges.

These were carefully selected and encoded, and redundant features were removed.




```python

# 1. Start from encoded data
df_fe = df_encoded.copy()

# 2. Feature: Log-transformed training_hours
df_fe['training_hours_log'] = np.log1p(df_fe['training_hours'])

# 3. Feature: Interaction between dev index and training hours
df_fe['devindex_training_interaction'] = df_fe['city_development_index'] * df_fe['training_hours']

# 4. Group rare company sizes â€”
rare_sizes = df_fe['company_size'].value_counts()[df_fe['company_size'].value_counts() < 200].index
df_fe['company_size_grouped'] = df_fe['company_size'].replace(rare_sizes, 'Other')

# 5. Label encode new categorical column
if 'company_size_grouped' in df_fe.columns:
     df_fe['company_size_grouped'] = LabelEncoder().fit_transform(df_fe['company_size_grouped'])

# 6. Final check
df_fe.head()

```

#### 5.6.1 | Save LabelEncoder for company_size_grouped

```python

# Fit on original string labels
le_company_size_grouped = LabelEncoder()
df_fe['company_size_grouped'] = le_company_size_grouped.fit_transform(df_fe['company_size_grouped'])

# Save the encoder
joblib.dump(le_company_size_grouped, '../encoders/le_company_size_grouped.pkl')

# Optional: print classes to confirm
print("Classes:", le_company_size_grouped.classes_)


```

After performing feature engineering (e.g., combining features, transforming values), now apply SMOTE (Synthetic Minority Oversampling Technique) on the final dataset 
`df_fe` to balance the target classes.

SMOTE helps generate synthetic samples for the minority class, which improves model 
performance especially when data is imbalanced.

Step 5.7 | Handle Class Imbalance (SMOTE on Feature Engineered Data)

After feature engineering, SMOTE is applied again on the updated dataset `df_fe` to balance the target classes. This is necessary because feature engineering changes the structure of the data, and SMOTE must be applied afterward to match the final features.



```python

# 1. Define final features and target
X_fe = df_fe.drop('target', axis=1)
y_fe = df_fe['target']

# 2. Train-test split
X_train_fe, X_test_fe, y_train_fe, y_test_fe = train_test_split(
    X_fe, y_fe, test_size=0.2, stratify=y_fe, random_state=42
)

# Check class distribution before SMOTE
print("Before SMOTE:")
print(y_train_fe.value_counts())
```

```python
# 3. Apply SMOTE on training set only
smote = SMOTE(random_state=42)
X_train_fe_balanced, y_train_fe_balanced = smote.fit_resample(X_train_fe, y_train_fe)

# 4. Check class distribution after SMOTE
print("After SMOTE:\n", y_train_fe_balanced.value_counts())
```

## Step 6 | Model Building & Evaluation



### 6.1 | Logistic Regression

A Logistic Regression model is trained using a pipeline with a StandardScaler. This makes the code modular, clean, and easier to manage.

**Steps included:**
- Scaling the features using StandardScaler
- Training the Logistic Regression model
- Predicting on test set
- Evaluating using accuracy, ROC-AUC, classification report
- Visualizing the confusion matrix using Seaborn


```python

# 1. Define the Pipeline
lr_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lr', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42))
])
# 2. Fit the model 
lr_pipe.fit(X_train_fe_balanced, y_train_fe_balanced)

# 3. Make predictions
y_pred_lr = lr_pipe.predict(X_test_fe)

# 4. Evaluation metrics
print("Logistic Regression Classification Report:")
print(classification_report(y_test_fe, y_pred_lr))
print("Accuracy Score:", accuracy_score(y_test_fe, y_pred_lr))
print("ROC-AUC Score:", roc_auc_score(y_test_fe, y_pred_lr))
ConfusionMatrixDisplay.from_predictions(y_test_fe, y_pred_lr, cmap='Greens')
```

 Note: The accuracy of Logistic Regression dropped slightly after applying feature engineering.

This is expected because:

Logistic Regression is a linear model, and not all engineered features added linear separability.

Some of the new features may not have high predictive value and could increase noise.

Many new columns were added, but didnâ€™t perform feature selection, which can impact simpler models.

It was chosen not to further optimize Logistic Regression, as:

Itâ€™s included for comparison and baseline purposes only.

The focus is on building better-performing tree-based models like Random Forest, XGBoost, and LightGBM which handle complex features better.


### 6.2 | Random Forest


In this step, a Random Forest Classifier was built to predict employee retention.  
To enhance model performance,  **GridSearchCV** was used to automatically find the best hyperparameter combinations. This ensures the model is optimized rather than relying on default settings.

evaluation done using:
- Confusion Matrix
- Classification Report
- Accuracy Score
- ROC-AUC Score


```python

# 1. Define the Pipeline
rf_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('rf', RandomForestClassifier(random_state=42))
])
# 2. Define hyperparameter grid
rf_params = {
    'rf__n_estimators': [100, 200],
    'rf__max_depth': [10, None],
    'rf__min_samples_split': [2, 5],
    'rf__min_samples_leaf': [1, 2],
    'rf__max_features': ['sqrt']
}

# 3. Set up GridSearchCV
rf_grid = GridSearchCV(rf_pipe, rf_params, scoring='f1_weighted', cv=3, n_jobs=-1, verbose=2)

# 4. Fit model on SMOTE-balanced training data
rf_grid.fit(X_train_fe_balanced, y_train_fe_balanced)

# 5. Best model
best_rf_model = rf_grid.best_estimator_

# 6. Predict on test set
y_pred_rf = rf_grid.predict(X_test_fe)

# Evaluation
print("Random Forest Classification Report:")
print(classification_report(y_test_fe, y_pred_rf))
print("Accuracy Score:", accuracy_score(y_test_fe, y_pred_rf))
print("ROC-AUC Score:", roc_auc_score(y_test_fe, y_pred_rf))
ConfusionMatrixDisplay.from_predictions(y_test_fe, y_pred_rf, cmap='Greens')
```

### 6.3 | XGBoost

XGBoost pipeline with:
- Refined features from error analysis.
- Feature selection using feature importance.
- Hyperparameter tuning via GridSearchCV.

Final Accuracy: **~76%**, ROC-AUC: **~0.744**

- Best generalization and recall among all models  


```python

from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, ConfusionMatrixDisplay

# Create pipeline
xgb_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('xgb', XGBClassifier(random_state=42, use_label_encoder=False, eval_metric='logloss'))
])

# Define parameter grid
param_grid = {
    'xgb__n_estimators': [100, 200],
    'xgb__learning_rate': [0.05, 0.1],
    'xgb__max_depth': [5, 7],
    'xgb__subsample': [0.8, 1.0]
}

# Grid Search
grid_search = GridSearchCV(xgb_pipe, param_grid, cv=3, scoring='f1_weighted', n_jobs=-1, verbose=2)
grid_search.fit(X_train_fe_balanced, y_train_fe_balanced)

# Best estimator
best_xgb = grid_search.best_estimator_

# Predictions
y_pred = best_xgb.predict(X_test_fe)

# Evaluation
print("\n\033[1mClassification Report:\033[0m")
print(classification_report(y_test_fe, y_pred))
print("\033[1mAccuracy Score:\033[0m", accuracy_score(y_test_fe, y_pred))
print("\033[1mROC-AUC Score:\033[0m", roc_auc_score(y_test_fe, y_pred))
ConfusionMatrixDisplay.from_predictions(y_test_fe, y_pred, cmap='Greens')

```

### 6.4 | LightGBM

```python
# Build pipeline
lgbm_pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('lgbm', LGBMClassifier(random_state=42))
])

# Parameter grid
lgbm_params = {
    'lgbm__n_estimators': [100, 200],
    'lgbm__learning_rate': [0.05, 0.1],
    'lgbm__max_depth': [5, 10],
    'lgbm__num_leaves': [31, 50]
}

# Grid search
lgbm_grid = GridSearchCV(lgbm_pipe, lgbm_params, scoring='f1_weighted', cv=3, n_jobs=-1, verbose=2)
lgbm_grid.fit(X_train_fe_balanced, y_train_fe_balanced)

# 5. Best model
best_lgbm = lgbm_grid.best_estimator_

#6. Predict
y_pred_lgbm = lgbm_grid.predict(X_test_fe)

#7. Evaluation
print("LightGBM Classification Report:")
print(classification_report(y_test_fe, y_pred_lgbm))
print("Accuracy Score:", accuracy_score(y_test_fe, y_pred_lgbm))
print("ROC-AUC Score:", roc_auc_score(y_test_fe, y_pred_lgbm))
ConfusionMatrixDisplay.from_predictions(y_test_fe, y_pred_lgbm, cmap='Greens')
```

### 6.5 | Final Model â€“ Voting Ensemble (Soft Voting)

After evaluating individual models:

- **Logistic Regression**: Good baseline, but lower recall for Class 1.
- **Random Forest**: Improved accuracy but still struggled with minority class.
- **XGBoost**: Strong performance, especially after hyperparameter tuning and error analysis.
- **LightGBM**: Highest standalone accuracy (78.3%) and better recall on Class 1.

To combine strengths of different models, a **soft voting ensemble** was built using:
- XGBoost
- LightGBM
- Random Forest

Different weight combinations were tested to favor stronger learners, especially LGBM.  
**Final weight configuration**: `[0, 4, 1]` (LGBM gets more influence)

This gave the best overall performance:

- **Accuracy:** 78.4%
- **Recall (Class 1):** 63%
- **ROC-AUC:** 0.734

This balanced both classes better and gives a robust final model for deployment.


```python

# Final voting ensemble with higher weight to LightGBM
voting_clf_final = VotingClassifier(
    estimators=[
        ('xgb', best_xgb),
        ('lgbm', best_lgbm),
        ('rf', best_rf_model)
    ],
    voting='soft',
    #weights=[1, 2, 1],  # LGBM gets more influence
    #weights=[1, 3, 1],
    #weights=[0, 3, 1],
    weights=[0, 4, 1],  #Final Voting Ensemble
    n_jobs=-1
)

# Fit on full training data
voting_clf_final.fit(X_train_fe_balanced, y_train_fe_balanced)

# Predict on test
y_pred_vote_final = voting_clf_final.predict(X_test_fe)

# Evaluate
print("\033[1mVoting Ensemble (LGBM-weighted) Classification Report:\033[0m")
print(classification_report(y_test_fe, y_pred_vote_final))

print("\033[1mAccuracy Score:\033[0m", accuracy_score(y_test_fe, y_pred_vote_final))
print("\033[1mROC-AUC Score:\033[0m", roc_auc_score(y_test_fe, y_pred_vote_final))

# Plot confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test_fe, y_pred_vote_final, cmap='Greens')
plt.title("Voting Ensemble Confusion Matrix")
plt.show()
```
